Q7: Three resiters of 10, 20 and 30 ohms are connected in series and in parallel if 10 V is given to them find current and voltages across them.

```python
#Time:20min
#In series circuit
R1=10
R2=20
R3=30
R=R1+R2+R3
print("The total resistance in series:",R,"ohms")
V=10
I=V/R
print("The current in series resistors is:",I,"amps")
#VOLTAGE ACROSS R1
print("The volatge across R1:",I*R1,"Volts")
#VOLTAGE ACROSS R2
print("The voltage across R2:",I*R2,"Volts")
#VOLATGE ACROSS R3
print("The voltage across R3:",I*R3,"Volts")
```

```python
#In parallel circuit
R1=10
R2=20
R3=30
R=((1/R1)+(1/R2)+(1/R3))**-1
print("The total resistance in parallel is:",R,"ohms")
V=10
print("The voltage in parallel combination is:",V,"volts")
V1=V2=V3=V
I1=V1/R1
I2=V2/R2
I3=V3/R3
#CURRENT ACROSS R1
print("The current across R1 is:",I1,"amps")
#CURRENT ACROSS R2
print("The current across R2 is:",I2,"amps")
#CURRENT ACROSS R3
print("The current across R3 is:",I3,"amps")

```
